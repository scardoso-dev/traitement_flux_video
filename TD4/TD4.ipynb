# Rapport du TD4 : Détection de Véhicules et Suivi de Vitesse

## Objectif
L'objectif principal de ce TD est de détecter des véhicules dans une vidéo et de suivre leur trajectoire en estimant leur vitesse. Cela a été réalisé à l'aide des bibliothèques **OpenCV**, **dlib**, et **IPython** pour une intégration dans un environnement Jupyter Notebook.

---

## Méthodologie

### Chargement des Fichiers
- **Modèle Haar Cascade :** Utilisation de `myhaar.xml` pour détecter les véhicules.
- **Vidéo :** Lecture de `cars.mp4` pour analyser le trafic.

### Détection des Véhicules
- Le modèle Haar Cascade détecte les véhicules dans chaque frame.
- La détection est effectuée toutes les 10 frames pour réduire la charge de calcul.

### Suivi des Véhicules
- Chaque véhicule détecté est suivi à l'aide du **Correlation Tracker** de la bibliothèque dlib.
- Des rectangles sont tracés autour des véhicules pour visualiser leur position.

### Estimation de la Vitesse
- La vitesse est calculée à partir des déplacements successifs des véhicules entre deux frames :
  - **Formule utilisée :**  
    \[
    \text{Vitesse} = \frac{\text{Distance (pixels)}}{\text{Pixels par mètre}} \times \text{FPS} \times 3.6
    \]
  - Les paramètres de calibration incluent :
    - Pixels par mètre (ppm) : 8.8.
    - Images par seconde (FPS) : 30.
